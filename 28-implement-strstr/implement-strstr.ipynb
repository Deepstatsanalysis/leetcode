```python
class Solution:
    def strStr(self, haystack, needle):
        """
        :type haystack: str
        :type needle: str
        :rtype: int
        """
        if not needle: return 0
        
        for i in range(len(haystack) - len(needle) + 1):
            if haystack[i:i+len(needle)] == needle:
                return i
                
        return -1
```

```python
s = Solution()
data = [
    ["aaaaa", ""],
    ["hello", "ll"],
    ["aaaaa", "bba"],
    [""
, ""],
#     ["", ""],
#     ["", ""],
]
for i in data:
    print(s.strStr(i[0], i[1]))
```
